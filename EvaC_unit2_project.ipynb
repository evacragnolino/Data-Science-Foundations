```python
!pip install preliz

import pymc as pm
import numpy as np
import preliz as pz
```

```python
import arviz as az
```

```python
WC_finals = np.repeat([1,0], repeats = [16,48])
```

```python
with pm.Model() as WC_finals_model:
  # prior from past World Athletics Finals
  θ = pm.Beta('θ' , alpha = 0.5 , beta = 4.5)

  #same likelihood, with our stand in for the binomial distribution
  #but with the updated data
  y = pm.Bernoulli('y', p = θ, observed = WC_finals)

  #sample from the posterior, 2000 for a more prcise picture later
  idata_WC_finals = pm.sample(2000)
```

```python
az.plot_posterior(idata_WC_finals)
```

```python
pz.Beta(alpha=1,beta=9).plot_pdf()
```
